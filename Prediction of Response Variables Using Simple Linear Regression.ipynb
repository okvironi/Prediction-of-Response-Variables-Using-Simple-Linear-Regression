```python
#SAMPLE PROGRAM 1

import numpy as np
import matplotlib.pyplot as plt
from numpy import array
from numpy.linalg import inv

class Metrics:
  def sse(self, Y, pred): #tambahkan argumen Y dan pred
    squared_error = (Y-pred)**2
    self.sse = np.sum(squared_error)
    return self.sse


class RegresiLinear:
  def __init__(self, array_input):
    self.prediction = None
    self.coefficient = None
    self.array_input = array_input

  def fitting(self, X, Y):
    self.X = self.array_input #ganti ke argumen array input
    self.Y = Y
    self.coefficient = inv(self.X.T.dot(self.X)).dot(self.X.T).dot(self.Y)


#menampilkan hasil regresi
if __name__ == "__main__":
  #data input (X) dan variabel dependen (Y)
  #X = array ([[1, 2], [2,3], [3,4]])
  #Y = array([2, 4, 6])


  #test manual input data
  #Meminta jumlah data
  n = int(input("Masukkan jumlah data: "))

  # Meminta input data X dan Y
  X = []
  Y = []

  for i in range(n):
    x_val = float(input(f"Masukkan nilai X-{i + 1}: "))
    y_val = float(input(f"Masukkan nilai Y-{i + 1}: "))
    X.append([x_val])
    Y.append([y_val])

  X = array(X)
  Y = array(Y)


  #inisialisasi model regresi linear
  model = RegresiLinear(X)

  #Melakukan fitting model
  model.fitting(X, Y)

  #Melakukan prediksi
  model.pred = X.dot(model.coefficient)

  #Menampilkan hasil prediksi
  print("\n" + "Hasil Prediksi:")
  print(model.pred, "\n")

  #Menampilkan koefisien model
  print("Koefisien Model:")
  print(model.coefficient, "\n")

  #Menghitung dan menampilkan SSE
  metrics = Metrics()
  sse = metrics.sse(Y, model.pred)
  print("SSE:", sse, "\n")


  #menampilkan grafik hasil regresi
  plt.scatter(X[:, 0], Y, label='Data Aktual')
  plt.plot(X[:, 0], model.pred, color='red', label='Regresi Linear')
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.legend()
  plt.show()
```

```python
#SAMPLE PROGRAM 2
#ERROR R-SQUARED

import numpy as np
import matplotlib.pyplot as plt
from numpy import array
from numpy.linalg import inv


class Metrics:
  def sse(self, Y, pred): #tambahkan argumen Y dan pred
    squared_error = (Y-pred)**2
    self.sse_result = np.sum(squared_error) #untuk menghidari konflik dengan MSR
    return self.sse_result

  def mse(self, Y, pred):
    n = len(Y)
    mse = self.sse(Y, pred) / n
    return mse

  def r_squared(self, Y, pred):
    sse = self.sse(Y, pred)
    sst = np.sum((Y - np.mean(Y))**2)
    r2 = 1 - (sse / sst)
    return r2


class RegresiLinear:
  def __init__(self, array_input):
    self.prediction = None
    self.coefficient = None
    self.array_input = array_input

  def fitting(self, X, Y):
    self.X = self.array_input #ganti ke argumen array input
    self.Y = Y
    self.coefficient = inv(self.X.T.dot(self.X)).dot(self.X.T).dot(self.Y)


#menampilkan hasil regresi
if __name__ == "__main__":
  #data input (X) dan variabel dependen (Y)
  #X = array ([[1, 2], [2,3], [3,4]])
  #Y = array([2, 4, 6])


  #test manual input data
  #Meminta jumlah data
  n = int(input("Masukkan jumlah data: "))

  # Meminta input data X dan Y
  X = []
  Y = []

  for i in range(n):
    x_val = float(input(f"Masukkan nilai X-{i + 1}: "))
    y_val = float(input(f"Masukkan nilai Y-{i + 1}: "))
    X.append([x_val])
    Y.append([y_val])

  X = array(X)
  Y = array(Y)


  #inisialisasi model regresi linear
  model = RegresiLinear(X)

  #Melakukan fitting model
  model.fitting(X, Y)

  #Melakukan prediksi
  model.pred = X.dot(model.coefficient)

  #Menampilkan hasil prediksi
  print("\n" + "Hasil Prediksi:")
  print(model.pred, "\n")

  #Menampilkan koefisien model
  print("Koefisien Model:")
  print(model.coefficient, "\n")

  #Menghitung dan menampilkan SSE
  metrics = Metrics()
  sse = metrics.sse(Y, model.pred)
  print("SSE:", sse, "\n")

  # Menghitung dan menampilkan MSE (means squared error)
  mse = metrics.mse(Y, model.pred)
  print("MSE:", mse, "\n")

  # Menghitung dan menampilkan R-squared
  r2 = metrics.r_squared(Y, model.pred)
  print("R-squared:", r2, "\n")

  #menampilkan grafik hasil regresi
  plt.scatter(X[:, 0], Y, label='Data Aktual')
  plt.plot(X[:, 0], model.pred, color='red', label='Regresi Linear')
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.legend()
  plt.show()
```

## **NOTE**

**PENGUJIAN HASIL**


1.   **Mean Squared Error (MSE)**: mengukur rata-rata dari kuadrat perbedaan antara nilai prediksi dan nilai aktual. Semakin kecil MSE, semakin baik model.
2.  **R-squared (R^2)**: mengukur sejauh mana variabilitas dalam data target (Y) dapat dijelaskan oleh model. Nilai R^2 berkisar antara 0 dan 1, di mana 1 berarti model Anda menjelaskan seluruh variabilitas dan 0 berarti model tidak menjelaskan variabilitas sama sekali.
3.   **Visualisasi Grafik**: menguji hasil model dengan memvisualisasikan hasil regresi. Jika garis regresi mengikuti tren data dengan baik, itu adalah tanda baik. Jika ada perbedaan besar antara data aktual dan garis regresi, maka model mungkin perlu ditingkatkan
